```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración para gráficos en español
plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['font.size'] = 10

def calcular_esfuerzo(S):
    """
    Calcula el esfuerzo basado en el tamaño del proyecto
    E = 8 * S^0.95
    """
    return 8 * (S ** 0.95)

def calcular_tiempo_calendario(E):
    """
    Calcula el tiempo calendario basado en el esfuerzo
    td = 2.4 * E^0.33
    """
    return 2.4 * (E ** 0.33)

# Función principal para calcular esfuerzo y tiempo dado un tamaño
def calcular_proyecto(S):
    """
    Función principal que calcula esfuerzo y tiempo calendario
    """
    E = calcular_esfuerzo(S)
    td = calcular_tiempo_calendario(E)
    
    print(f"Tamaño del proyecto (S): {S}")
    print(f"Esfuerzo calculado (E): {E:.2f}")
    print(f"Tiempo calendario (td): {td:.2f}")
    
    return E, td

# Ejemplo de uso con un proyecto específico
print("=== Ejemplo de cálculo para un proyecto ===")
S_ejemplo = 1000
E_ejemplo, td_ejemplo = calcular_proyecto(S_ejemplo)

print("\n" + "="*50)
print("GENERACIÓN DE GRÁFICAS")
print("="*50)

# 1. Gráfica de Esfuerzo (E) para tamaños [0, 10000]
print("\n1. Calculando valores de Esfuerzo para tamaños [0, 10000]...")
tamaños = np.linspace(1, 10000, 1000)  # Evitamos 0 para evitar problemas con exponentes
esfuerzos = calcular_esfuerzo(tamaños)

# 2. Gráfica de tiempo calendario (td) para esfuerzos [1, 500]
print("2. Calculando valores de Tiempo calendario para esfuerzos [1, 500]...")
esfuerzos_rango = np.linspace(1, 500, 500)
tiempos_calendario = calcular_tiempo_calendario(esfuerzos_rango)

# Crear las gráficas
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Gráfica 1: Esfuerzo vs Tamaño
ax1.plot(tamaños, esfuerzos, 'b-', linewidth=2, label='E = 8 * S^0.95')
ax1.set_xlabel('Tamaño del proyecto (S)')
ax1.set_ylabel('Esfuerzo (E)')
ax1.set_title('Esfuerzo vs Tamaño del Proyecto')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Marcar el punto del ejemplo
E_ejemplo_graf = calcular_esfuerzo(S_ejemplo)
ax1.plot(S_ejemplo, E_ejemplo_graf, 'ro', markersize=8, label=f'Ejemplo: S={S_ejemplo}')
ax1.legend()

# Gráfica 2: Tiempo calendario vs Esfuerzo
ax2.plot(esfuerzos_rango, tiempos_calendario, 'r-', linewidth=2, label='td = 2.4 * E^0.33')
ax2.set_xlabel('Esfuerzo (E)')
ax2.set_ylabel('Tiempo calendario (td)')
ax2.set_title('Tiempo Calendario vs Esfuerzo')
ax2.grid(True, alpha=0.3)
ax2.legend()

# Marcar el punto del ejemplo
td_ejemplo_graf = calcular_tiempo_calendario(E_ejemplo_graf)
ax2.plot(E_ejemplo_graf, td_ejemplo_graf, 'ro', markersize=8, label=f'Ejemplo: E={E_ejemplo_graf:.1f}')
ax2.legend()

plt.tight_layout()
plt.show()

# Tabla de valores de ejemplo
print("\n" + "="*50)
print("TABLA DE VALORES DE EJEMPLO")
print("="*50)
print(f"{'Tamaño (S)':<12} {'Esfuerzo (E)':<15} {'Tiempo (td)':<15}")
print("-" * 45)

tamaños_ejemplo = [100, 500, 1000, 2000, 5000, 8000, 10000]
for S in tamaños_ejemplo:
    E = calcular_esfuerzo(S)
    td = calcular_tiempo_calendario(E)
    print(f"{S:<12} {E:<15.2f} {td:<15.2f}")

# Análisis adicional
print("\n" + "="*50)
print("ANÁLISIS DE LAS RELACIONES")
print("="*50)
print("1. La relación E = 8 * S^0.95 muestra que el esfuerzo crece más que linealmente")
print("   con el tamaño (exponente > 1), indicando economías de escala negativas.")
print("\n2. La relación td = 2.4 * E^0.33 muestra que el tiempo calendario crece")
print("   menos que proporcionalmente al esfuerzo (exponente < 1),")
print("   sugiriendo que proyectos más grandes pueden paralelizarse mejor.")

# Verificación de fórmulas
print(f"\n3. Para el ejemplo S={S_ejemplo}:")
print(f"   - Esfuerzo: E = 8 * {S_ejemplo}^0.95 = {E_ejemplo:.2f}")
print(f"   - Tiempo: td = 2.4 * {E_ejemplo:.2f}^0.33 = {td_ejemplo:.2f}")
```
